```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.8
b = -0.0009
T = 1000

c_max = 4.
c_min = 1.

all_t = np.arange(T)

x = np.zeros(T)

for t in range(1, T):
    x[t] = b * all_t[t] + a * x[t-1] + np.random.normal(0, 1)

x =  (x - min(x)) / (max(x) - min(x)) * (c_max - c_min) + c_min

plt.plot(x)
plt.show()
```

```python
# Checking to see how to retrive the exact process that you want from the cost 
# All of this here is for the integration of the cost 

np.random.seed(50)
J = 2
a = 0.8
b = -0.0009
T = 10  # Increased T value

c_max_scale = 4.
c_min_scale = 1.

C = np.zeros((J, T))
for t in range(1, T):
    C[:, t] = b * t + a * C[:, t-1] + np.random.normal(0, 1, size=J)

print(C)
c_max = np.tile(np.array([np.amax(C, axis=1)]).T, (1, T))
print(c_max)
c_min = np.tile(np.array([np.amin(C, axis=1)]).T, (1, T))
print(c_min)


C = (C - c_min) / (c_max - c_min) * (c_max_scale - c_min_scale) + c_min_scale
print(C)
```
