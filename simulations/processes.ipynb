```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.9
b = 0.004
T = 1000

c_max = 10.
c_min = 1.

all_t = np.arange(T)

x = np.zeros(T)

for t in range(1, T):
    x[t] = b * all_t[t] + a * x[t-1] + np.random.normal(0, 1)

x =  (x - min(x)) / (max(x) - min(x)) * (c_max - c_min) + c_min

plt.plot(x)
plt.show()
```
