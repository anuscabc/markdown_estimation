```python
import numpy as np
from scipy.stats import norm

# Set the number of dimensions and the number of samples
n_dims = 99
n_samples = 500

# Generate a Halton sequence using numpy
from numpy.random import default_rng
rng = default_rng(123) # set the random seed for reproducibility
halton = rng.random((n_samples, n_dims))

# Transform the Halton sequence to a standard normal distribution using the inverse CDF of the normal distribution
samples = norm.ppf(halton)
print(samples.shape)
# Print the mean and variance of the samples
print(f"Mean: {samples.mean(axis=0)}")
print(f"Variance: {samples.var(axis=0)}")
```

```python
import numpy as np
from scipy.stats import gumbel_r

# Set the number of dimensions and the number of samples
n_dims = 99
n_samples = 500

# Set the location and scale parameters of the Gumbel distribution
mu = 0
beta = 1

# Generate a Halton sequence using numpy
from numpy.random import default_rng
rng = default_rng(123) # set the random seed for reproducibility
halton = rng.random((n_samples, n_dims))

# Transform the Halton sequence to the Gumbel distribution using the inverse CDF
samples = mu - beta * np.log(-np.log(halton))

# Print the mean and variance of the samples
print(f"Mean: {samples.mean(axis=0)}")
print(f"Variance: {samples.var(axis=0)}")
```
