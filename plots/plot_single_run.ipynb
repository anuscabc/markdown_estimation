```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/market_100.csv')
```

```python
df_grouped = df.groupby('product')
print(df_grouped)
# Look at the firm with bad product charactristics
group1 = df_grouped.get_group(1)
# Look at a firm with good product characteristics
group4 = df_grouped.get_group(4)

```

```python
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2, 2)

for firm in df['product'].unique():
    df_firm = df[df['product'] == firm]
    ax1.plot(df_firm.time, df_firm.price)
    ax2.plot(df_firm.time, df_firm.cost)
    ax3.plot(df.time, df.markups)
    ax4.plot(df_firm.time, df_firm.mshare)

ax1.set_title('price')
ax2.set_title('cost')
ax3.set_title('markups')
ax4.set_title('market share')

plt.tight_layout()
plt.show()
```

```python
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2, 2)
fig.suptitle(f'Firm with bad product characteristics')

ax1.plot(group1.time, group1.price)
ax2.plot(group1.time, group1.cost)
ax3.plot(group1.time, group1.markups)
ax4.plot(group1.time, group1.mshare)

ax1.set_title('price')
ax2.set_title('cost')
ax3.set_title('markups')
ax4.set_title('market share')

plt.tight_layout()
plt.show()

```

```python
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2, 2)
fig.suptitle(f'Firm with good product characteristics')

ax1.plot(group4.time, group4.price, color='orange')
ax2.plot(group4.time, group4.cost, color='orange')
ax3.plot(group4.time, group4.markups, color='orange')
ax4.plot(group4.time, group4.mshare, color='orange')

ax1.set_title('price')
ax2.set_title('cost')
ax3.set_title('markups')
ax4.set_title('market share')

plt.tight_layout()
plt.show()
```
